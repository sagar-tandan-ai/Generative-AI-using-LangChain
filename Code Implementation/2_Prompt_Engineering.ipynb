## Prompt Engineering

**First we need to load the model, we have already discussed how we can load models in previous session**

```python
import os
from dotenv import load_dotenv 

load_dotenv(override=True)

GROQ_API_KEY = os.getenv("GROQ_API_KEY")
GROQ_BASE_URL = "https://api.groq.com/openai/v1"
```

```python
from langchain_openai import ChatOpenAI

model = ChatOpenAI(
    model="llama-3.1-8b-instant", 
    api_key=GROQ_API_KEY, 
    base_url=GROQ_BASE_URL
)
```

### 1. Zero-Shot Prompting

```python
zero_shot_prompt = """
Classify the sentiment of these reviews in one word as:
"Positive", "Negative", "Neutral"
Here is the sentence: 
"The hotel was okay. Room was clean but service was slow."  
"""
```

```python
response = model.invoke(zero_shot_prompt)
print(response.content)
```

### 2. One-Shot Prompting 

```python
one_shot_prompt = """
Example: 
English: Hello
Nepali: Namaste
Now translate: 
English: Good Morning 
""" 
print(one_shot_prompt)
```

```python
response = model.invoke(one_shot_prompt)
print(response.content)
```

### 3. Few-Shot Prompting

```python
few_shot_prompt = """
Convert each sentence into a suitable emoji. Here are some examples: 
Sentence: I got 4 GPA in high school exams.
Emoji: 🎉📚
Sentence: It's raining outside. 
Emoji: 🌧️☔ 
Sentence: I am in love with a girl.
Emoji: 💘👧
Sentence: I am going to propose to a girl for marriage. 
Emoji: 
""" 
print(few_shot_prompt)
```

```python
response = model.invoke(few_shot_prompt)
print(response.content)
```

### 4. Chain-of-Thought (CoT) Prompting

```python
from IPython.display import Markdown

# without CoT: 
no_cot_prompt ="""
Question: 
If 6 printers can print 6 pages in 6 minutes, 
how many pages can 12 printers print in 12 minutes?

Provide only the final answer.
"""
response = model.invoke(no_cot_prompt)
display(Markdown(response.content)) 
```

```python
# with CoT Prompting
cot_prompt = """
Question:
If 6 printers can print 6 pages in 6 minutes,
how many pages can 12 printers print in 12 minutes?

Think through the problem carefully before answering.
After reasoning, provide:
1. A brief solution.
2. The final answer.
"""
response = model.invoke(cot_prompt)
display(Markdown(response.content)) 
```

### 5. Role Prompting

```python
# Without role prompting
prompt = "Explain playing Card in short."
response = model.invoke(prompt)
display(Markdown(response.content))
```

```python
# With role prompting
prompt = """
You are a math Professor who excels in probability. 
Explain playing cards in short.
"""
response = model.invoke(prompt)
display(Markdown(response.content))
```

### 6. Instruction Prompting

```python
instruction_prompting = """
Summarize the following article.

Instructions:
- Use exactly 4 bullet points.
- Keep each bullet under 15 words.
- Do not include opinions.
- Focus only on the key facts.
- Please write it in proper markdown 

{article}
"""
```

```python
# reading the file for {article}
with open("../1_Datasets/article.txt", "r", encoding="utf-8") as f: 
    article = f.read()
print(article)
```

```python
# sending the prompt with article to LLM 
response = model.invoke(instruction_prompting.format(article=article))
display(Markdown(response.content))
```
