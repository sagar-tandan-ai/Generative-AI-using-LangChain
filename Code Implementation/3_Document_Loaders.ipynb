### 1. TextLoader 

```python
# importing the TextLoader library from langchain
from langchain_community.document_loaders import TextLoader

loader = TextLoader(file_path="../1_Datasets/Harry_Potter.txt", encoding="utf-8")
docs = loader.load()
```

```python
print("The data type of the document:", type(docs))
print("The length of documents (number of total documents):", len(docs))
print(docs)
```

```python
print("Metadata of documents: \n", docs[0].metadata)
print("\nText content: \n", docs[0].page_content[:200])
```

### 3. WebBaseLoader

```python
#For this we need extra module called bs4
!uv add bs4 -q
```

```python
from langchain_community.document_loaders import WebBaseLoader

web_loader = WebBaseLoader(web_path="https://en.wikipedia.org/wiki/Harry_Potter")
Web_docs = web_loader.load()
```

```python
print("The data type of the document:", type(Web_docs))
print("The length of documents (number of total documents):", len(Web_docs))
print(Web_docs)
```

```python
print("Metadata of documents: \n", Web_docs[0].metadata)
# print("\nText content: \n", Web_docs[0].page_content[:200])
```

```python
## Let's take this a step further using LLM
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI

load_dotenv()

OPENROUTER_API_KEY = os.getenv("OPENROUTER_API_KEY")
OPENROUTER_BASE_URL = "https://openrouter.ai/api/v1"

openai = ChatOpenAI(
    model = "openrouter/owl-alpha", # using the free model  
    api_key=OPENROUTER_API_KEY,
    base_url=OPENROUTER_BASE_URL
)
```

```python
# Making the prompt ready 
prompt = """
Answer the following questions: 
{questions}
based on the provided context: 
{context}
"""
```

```python
# Asking LLM to answer the question 
response = openai.invoke(prompt.format(
    questions = "When was Harry Potter written?", 
    context = Web_docs[0].page_content)).content
```

```python
# Lets display the response
from IPython.display import Markdown
display(Markdown(response))
```

### 4. CSVLoader

```python
from langchain_community.document_loaders import CSVLoader

csv_loader = CSVLoader(file_path="../1_Datasets/top_10_countries.csv", encoding="utf-8")
csv_docs = csv_loader.load()
```

### 5. DirectoryLoader (Final LORD)

```python
from langchain_community.document_loaders import DirectoryLoader, PyPDFLoader

dir_loader = DirectoryLoader(
    path="../1_Datasets/sample_pdf/", 
    glob="*.pdf", 
    loader_cls=PyPDFLoader # Need to mention the loader_cls here
)
dir_docs = dir_loader.load()
```

```python
print("The data type of the document:", type(dir_docs))
print("The length of documents (number of total documents):", len(dir_docs))
# print(dir_docs)
```

```python
# Lazy_load() -> Lazy loading returns the generator function 
lazy_dir_loader = DirectoryLoader(
    path="../1_Datasets/sample_pdf/",
    glob="*.pdf", 
    loader_cls=PyPDFLoader
)
gen_docs = lazy_dir_loader.lazy_load()
```

```python
for doc in gen_docs: 
    print(doc.metadata)
```
